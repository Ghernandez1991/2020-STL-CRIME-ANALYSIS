```python
#create a function to handle the data joins for all the datasets we have available
def datajoin(datafile, codesfile):
    codes_df = pd.read_csv(codesfile)
    crime_df = pd.read_csv(datafile, encoding='cp1252')
    joined_df = crime_df.join(codes_df, on='Neighborhood',  how='left', lsuffix='Crime',  rsuffix='Code')
    return joined_df
    
```

```python
#call each function for each month to do the join 
jan_df = datajoin('Data/January2020.csv', 'Data/STL_CODES.csv')
feb_df = datajoin('Data/February2020.csv', 'Data/STL_CODES.csv')
mar_df = datajoin('Data/March2020.csv', 'Data/STL_CODES.csv')
apr_df = datajoin('Data/April2020.csv', 'Data/STL_CODES.csv')
may_df = datajoin('Data/May2020.csv', 'Data/STL_CODES.csv')
june_df = datajoin('Data/June2020.csv', 'Data/STL_CODES.csv')
july_df = datajoin('Data/July2020.csv', 'Data/STL_CODES.csv')
aug_df = datajoin('Data/August2020.csv', 'Data/STL_CODES.csv')
sept_df = datajoin('Data/September2020.csv', 'Data/STL_CODES.csv')
oct_df = datajoin('Data/October2020.csv', 'Data/STL_CODES.csv')
nov_df = datajoin('Data/November2020.csv', 'Data/STL_CODES.csv')
dec_df = datajoin('Data/December2020.csv', 'Data/STL_CODES.csv')
```

```python
#create seperate dataframe lists for later
df_list = [jan_df,feb_df,mar_df,apr_df,may_df,june_df,july_df,aug_df,sept_df,oct_df,nov_df,dec_df ]
df_list2 = [feb_df,mar_df,apr_df,may_df,june_df,july_df,aug_df,sept_df,oct_df,nov_df,dec_df ]
```

```python
#see the length for each dataframe in the list, add it to the list len_list

len_list = []

for item in df_list:
    string_len = len(item)
    len_list.append(string_len)
    
 #convert the list of strings(of length) to integers   
list_ints = list(map(int, len_list))
#print the sum of those lists. 
print(sum(list_ints))
```

```python
#concat all the DF's togther for a master list of 2020
final_2df = pd.concat(df_list)
```

```python
#export to csv
final_2df.to_csv('Data/2020_STL_CRIME.csv')
```
